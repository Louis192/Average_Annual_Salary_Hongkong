I intend to use the academic_year field as part of the predictor variables.However this can only be achieved after 
I transform the feature

```python
# Create a new columns called converted year
df['converted'] = df['Academic_Year'].str.replace(r'/', '')
# Extract the first 4 digits from converted year field to represent year
df['year'] = df.converted.astype(str).str[:4].astype(int)
# Extract the last two digits from converted year field
df['day'] = df.converted.astype(str).str[-2:].astype(int)
# Now drop two fields Academic_year and converted
df=df.drop(['Academic_Year','converted'],axis=1)
df
```

join the features with dummy variable into a dataframe

```python
# Drop two features because we have already obtained dummy variables from them
df=df.drop(['Level_of_Study','Broad_Academic_Programme_Category'],axis=1)
df
```

Seems I will be using linear regression algorithm it is important to note that there are some conditions to consider such as multicolinearity.
Multicolinearity is when independent variables are correlated with each other,and if this happen then it will create
problems to the model.Below I will test for multicolinearity using variance inflation factor(VIF).
VIF can be used to quantify correlation between two predictor variables and generally a good vif should be below 10.
Also it is worthy to note that we do not have any issue with dummy variable trap.


```python
# Test for multicolinearity,but first drop target variable
X=df.iloc[:,1:13] # here the target variable is dropped
vif_data=pd.DataFrame()
vif_data['features']=X.columns
vif_data['VIF']=[variance_inflation_factor(X.values,i) 
for i in range (len(X.columns))]
print(vif_data)
```

It can be observed that year and day are highly correlated,so I will have to drop one and in order to decide on which predictor variable to drop,I  will plot a correlation heatmap to select the predictor variable that is more correlated to the target variable.


```python
# Test for multicollinearity after obtaining dummy variables
```

```python
# Now lets decide between year and day features which one to drop
# Create a heatmap for better visualization
import seaborn as sns
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

From above heatmap we can see that year and day has same correlation with target variable,so drop day because it seems
more meaningful to use year instead of day as predictor variable.

```python
df=df.drop(['day'],axis=1)
```

```python
df
```

Now after dropping day feature we can observe that the vif of all predictor variables are below 10 and this is
acceptable.

```python
X=df.iloc[:,1:13]
vif_data=pd.DataFrame()
vif_data['features']=X.columns
vif_data['VIF']=[variance_inflation_factor(X.values,i) 
for i in range (len(X.columns))]
print(vif_data)
```

split the dataset to target feature and predictor features,where y is target feature and X is predictor features

```python
y=df.iloc[:,0]
X=df.iloc[:,1:13]
```

## Model Builder

### Build The first model


From the model notes below,it is observed that the Cond No=1.18e+06 which implies that there are some numerical
problems with the model and as such I will need to optimize the model and in order to optimize the model I will standardize the dataset so that each feature will have equal contribution to the prediction because as we can see the 'year' features has larger values when compared to other predictor values.


```python
unscaled=X
unscaled = sm.add_constant(unscaled)

# Note the difference in argument order
model = sm.OLS(y, unscaled).fit()
predictions = model.predict(unscaled) # make the predictions by the model

# Print out the statistics
model.summary()
```

In order to optimize the model,I will scale all the predictor variables so that each predictor variable can have equal 
contribution to the model.There are different modules that can be used to scale the predictor features but seems we 
are dealing with a linear regression problem it will be best to use the standard scaler module.

```python
# Now lets normalize the datasets so as to enable all features have equal chance of contribution
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
scaled_features=scaling.fit_transform(X)
scaled_features
```

```python
# After scaling transform back to a dataframe so that the indexes and features names can be retained
scaled_features_df = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)
```

```python
X=scaled_features_df
X
```

### Build second model

The second model is better than the first model because it has less condition number compared to the  first model and 
the reason is because large condition number signifies numerical problems or multicolinearity.


```python
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions from the model

# Print out the statistics
model.summary()
```

## Model Interpretation

#### Model Accuracy

From the above model output we can observe that the R-squared is 0.698,R-squared is one of the evaluation metrics that can be used to evaluate linear regression model.The Rsquared can be used to tell how well the model is making its predictions on a scale from 0-1 with 1 been the best and 0 been the least.

Average_Annual_Salary=271.7830+37.7042(year)-52.5411(Sub_degree)+9.1828(Taught_Postgraduate)-34.8715(Undergraduate)\
                      +14.6467(Business_and_Management)+12.8505(Education)+6.5334(Engineering_and_Technology)\
                      +36.1729(Medicine_Dentistry_and_Health)+0.6489(Sciences)+ 17.6163(Social_Sciences)

For each year,the Average annual salary can be increase by $37.7042,and for each sub-degree the average annual salary decreases by $52.5411,this logic is applicable to every other predictor variables in the model.
The R-squared is 0.698 or 69.8% which implies that the model explains  69.8% variability in the dependent variable.
The adjusted R-squared is 68.8% which implies that 68% of the predictor variables actually have influence on the 
dependent variable.
The P>|t| will help us to identify predictor features that are correlated with the dependent variable and so any feature whose P>|t| value is less than 0.05 has a strong correlation with the target variable.


Plot a visualization of predictions vs True values

```python
ax = sns.regplot(x=predictions, y=y)
ax.set_xlabel("Predictions", fontsize = 10)
ax.set_ylabel("Actual", fontsize = 10)
plt.show()
```
